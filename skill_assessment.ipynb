## ğŸ“ˆ Visual Skill Assessment

```python
# Create radar chart of skills
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Bar chart of current levels
colors = ['#ff4444' if x <= 3 else '#ffaa00' if x <= 6 else '#00aa44' if x <= 8 else '#ff6600' 
          for x in df_skills['Current_Level']]

ax1.barh(df_skills['Domain'], df_skills['Current_Level'], color=colors)
ax1.set_xlabel('Skill Level (1-10)')
ax1.set_title('ğŸš€ Your Current Skill Levels')
ax1.set_xlim(0, 10)
ax1.grid(axis='x', alpha=0.3)

# Market demand vs your level
ax2.scatter(df_skills['Market_Demand'], df_skills['Current_Level'], 
           s=100, c=df_skills['Experience_Years'], cmap='viridis', alpha=0.7)
ax2.plot([0, 10], [0, 10], 'r--', alpha=0.5, label='Perfect Match Line')
ax2.set_xlabel('Market Demand')
ax2.set_ylabel('Your Skill Level')
ax2.set_title('ğŸ’° Market Demand vs Your Skills')
ax2.legend()
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Identify your strengths
strengths = df_skills[df_skills['Current_Level'] >= 7]['Domain'].tolist()
print("\nğŸ”¥ YOUR EXPERT-LEVEL STRENGTHS:")
for strength in strengths:
    print(f"   âœ… {strength}")
```
